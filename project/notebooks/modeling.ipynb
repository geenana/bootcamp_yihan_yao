```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("../data/processed/engineered_features.csv", parse_dates=['date'], index_col='date')
df.head()
```

```python
df['lag_vix_1'] = df['vix'].shift(1)
df['lag_vix_5'] = df['vix'].shift(5)
df['roll_mean_vix_5'] = df['vix'].rolling(5).mean().shift(1)
df['roll_vol_vix_20'] = df['vix'].rolling(20).std().shift(1)

df['lag_rv_1'] = df['rv_21d_pct'].shift(1)
df['lag_rv_5'] = df['rv_21d_pct'].shift(5)
df['roll_mean_rv_5'] = df['rv_21d_pct'].rolling(5).mean().shift(1)
df['roll_vol_rv_20'] = df['rv_21d_pct'].rolling(20).std().shift(1)

# Target: next-day realized volatility
df['y_next_rv'] = df['rv_21d_pct'].shift(-1)

df_feat = df.dropna().copy()
```

```python
# Time-aware split 
cut = int(len(df_feat) * 0.8)
train, test = df_feat.iloc[:cut], df_feat.iloc[cut:]

features = [
    'lag_vix_1','lag_vix_5','roll_mean_vix_5','roll_vol_vix_20',
    'lag_rv_1','lag_rv_5','roll_mean_rv_5','roll_vol_rv_20'
]
X_tr, X_te = train[features], test[features]
y_tr, y_te = train['y_next_rv'], test['y_next_rv']
```

```python
reg = Pipeline([('scaler', StandardScaler()),('linreg', LinearRegression())])
reg.fit(X_tr, y_tr)
pred = reg.predict(X_te)

# Metrics 
mse = mean_squared_error(y_te, pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_te, pred)
print(f'RMSE: {rmse:.4f}, MAE: {mae:.4f}')

# Residual Analysis 
residuals = y_te - pred

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(y_te.index, residuals, label='Residuals')
plt.axhline(0, color='r', linestyle='--')
plt.legend()
plt.title('Residuals over Time')

plt.subplot(1,2,2)
plt.scatter(pred, residuals, alpha=0.5)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Prediction Error Plot")
plt.show()

# Forecast Visualization 
plt.figure(figsize=(10,5))
plt.plot(y_te.index, y_te, label='True')
plt.plot(y_te.index, pred, label='Predicted')
plt.legend()
plt.title('Forecasting Next-Day Realized Volatility')
plt.show()
```

### Interpretation 
- **Performance Metrics**: RMSE is 7.51 and MAE is 5.24, suggesting that on average, the model's forecasts deviate about 5 units from the true realized volatility, with larger errors penalized more heavily in RMSE. For volatility forecasting, these error magnitudes are reasonable given the inherent noise. 
- **Residual & Prediction Error Analysis**: The residuals mostly hover around 0 but show bursts of large deviations during volatility spikes; this indicates the model captures normal regimes well but struggles during periods of extreme market stress. In the prediction error plot, residuals spread wider as predicted volatility increases, suggesting heteroskedasticity. 
- **Risk-Aware Interpretation**: Tail risk underestimation, regime dependence, over-reliance risk 
- **Key Modeling Assupmtions**: Linearity, stationarity, independence of residuals, feature sufficiency, homoscedasticity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# Load dataset 
df = pd.read_csv("../data/processed/preprocessed_dataset.csv")

# Profile numeric and categorical columns 
print("Dataset info:")
df.info()

print ("\nMissing values:")
print(df.isna().sum())

num_cols = df.select_dtypes(include=[np.number]).columns
desc = df[num_cols].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
display(desc)
```

```python
# Plot distributions (histograms + boxplots) 
fig, axes = plt.subplots(2,2, figsize=(18,10))

sns.histplot(df['spx_close'], kde=True, ax=axes[0,0])
axes[0,0].set_title('SPX Close Distribution')

sns.histplot(df['vix'], kde=True, ax=axes[0,1]) 
axes[0,1].set_title('VIX Distribution')

sns.boxplot(x=df['spx_close'], ax=axes[1,0])
axes[1,0].set_title('SPX Close (Outliers)')

sns.boxplot(x=df['vix'], ax=axes[1,1])
axes[1,1].set_title('VIX (Outliers)')

plt.tight_layout()
plt.show()
```

```python
# Bivariate visuals (relationships)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.scatterplot(data=df, x='spx_close', y='vix', ax=axes[0], alpha=0.6)
axes[0].set_title('SPX Close vs VIX')

sns.lineplot(data=df, x='date', y='spx_close', ax=axes[1], label='SPX')
sns.lineplot(data=df, x='date', y='vix', ax=axes[1], label='VIX')
axes[1].set_title('SPX & VIX over Time')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
corr = df[num_cols].corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

## Insights & Assumptions 

### Top 3 insights
- SPX Close and VIX distributions are right-skewed, with occasional extreme spikes (market shocks).
- SPX Close and VIX are negatively correlated: as equity prices fall, volatility tends to rise.
- Time series plot shows structural breaks (for example, sharp drops or spikes) indicating crisis periods; these may dominate model training if not adjusted. 

### Assumptions & risks 
- Assumes missing values are random and not systematically tied to events.
- Treats extreme volatility events as genuine market shocks, not data entry errors.
- Non-normal distributions may bias linear models if not transformed. 

### Next steps before modeling (cleaning & features) 
- Impute missing values (median for continuous).
- Apply winsorization or z-score filtering for extreme outliers, expecially for VIX spikes.
- Consider log-transforming SPX Close and VIX to address skewness.
- Engineer features like daily returns, volatility ratios, or lag features for modeling.

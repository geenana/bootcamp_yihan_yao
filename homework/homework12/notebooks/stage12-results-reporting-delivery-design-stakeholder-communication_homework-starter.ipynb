```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
sns.set(style='whitegrid')
plt.rcParams['figure.dpi'] = 120
np.random.seed(101)
```

## Load Your Data
You can load your real results or use a generated synthetic dataset as fallback.

```python
data_path = Path('../data/final_results.csv')

if data_path.exists():
    df = pd.read_csv(data_path)
else:
    # Synthetic fallback dataset
    df = pd.DataFrame({
        'scenario': ['baseline','alt_impute','alt_outlier'],
        'return': [0.12, 0.11, 0.135],
        'volatility': [0.18, 0.185, 0.19],
        'sharpe': [0.56, 0.49, 0.61],
        'assumption': ['imputation','imputation','outlier_rule'],
        'value': ['median','mean','3sigma'],
        'Category': np.random.choice(['X','Y','Z'], 3),
        'MetricA': np.random.normal(75, 15, 3),
        'MetricB': np.random.normal(150, 30, 3),
        'Date': pd.date_range('2025-02-01', periods=3)
    })
df.head()
```

## Helper: Export Directory

```python
img_dir = Path('../deliverables/images')
img_dir.mkdir(parents=True, exist_ok=True)

def savefig(name):
    plt.tight_layout()
    plt.savefig(img_dir / name, dpi=300)
    print(f'Saved {name}')
```

## Chart 1: Riskâ€“Return Scatter

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='volatility', y='return', hue='scenario', s=80)
plt.title('Riskâ€“Return by Scenario')
plt.xlabel('Volatility')
plt.ylabel('Return')
savefig('risk_return.png')
plt.show()
```

## Chart 2: Return by Scenario (Bar Chart)

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='scenario', y='return')
plt.title('Return by Scenario')
savefig('return_by_scenario.png')
plt.show()
```

## Chart 3: MetricA Over Time (Line Chart)

```python
plt.figure(figsize=(7,5))
sns.lineplot(data=df, x='Date', y='MetricA', hue='Category', marker='o')
plt.title('MetricA Over Time by Category')
plt.xlabel('Date')
plt.ylabel('MetricA')
savefig('metricA_over_time.png')
plt.show()
```

- **Chart 1 takeaway:** The chart is a scatter plot showing volatility in x-axis and return in y-axis. The imputation by mean method has middle volatility and the lowest return. It assumes missing data are not systematically biased. 
- **Chart 2 takeaway:** The chart is a bar plot comparing returns of different scenarios. - **Chart 3 takeaway:** The chart is a line chart, though because of limited data, it only shows dots there. It compared MetricA over time across scenarios. MetricA rises for category Y (about 67) and rises more for category X (about 72) as time goes by.

```python
def autodraft_report(assumptions: pd.DataFrame):

    lines = []
    lines.append("# Model Report\n")

    # Executive Summary 
    lines.append("## Executive Summary\n")
    lines.append("- **Objective**: Translate model results into decisions; expose assumptions and sensitivity.") 

    # Charts with Interpretation 
    lines.append("\n## Charts with Interpretation\n")
    lines.append("Below are the key charts from the analysis:\n")

    lines.append("![](../deliverables/images/risk_return.png)\n")
    lines.append("*Chart 1: Risk-Return by Scenario*\n")
    lines.append("â†’ The chart is a scatter plot showing volatility in x-axis and return in y-axis. The volatility of the \
    outliers removing method, which is about 0.19, is the highest; but the return of it is also the highest, which is \
    about 0.135. The imputation by mean method has middle volatility and the lowest return. It assumes missing data are \
    not systematically biased.\n")

    lines.append("![](../deliverables/images/return_by_scenario.png)\n")
    lines.append("*Chart 2: Return by Scenario*\n")
    lines.append("â†’ The chart is a bar plot comparing returns of different scenarios. The outlier removing method has the \
    highest return, which is 0.135; the baseline has the second highest return, which is 0.12; and the imputation by mean \
    method has the lowest return, which is 0.11. This confirms the model is sensitive to data treatment choices. It assumes \
    that stability of results depends on data quality and standardize preprocessing is needed to avoid inconsistent outcomes.\n")

    lines.append("![](../deliverables/images/metricA_over_time.png)\n")
    lines.append("*Chart 3: MetricA Over Time by Category*\n")
    lines.append("â†’ The chart is a line chart, though because of limited data, it only shows dots there. It compared \
    MetricA over time across scenarios. MetricA rises for category Y (about 67) and rises more for category X (about 72) \
    as time goes by.\n")

    # Assumptions & Risks 
    lines.append("## Assumptions & Risks\n")
    lines.append("- Assumes missing data are random and not systematically biased.\n")
    lines.append("- Assumes extreme outliers represent noise rather than meaningful stress events.\n")
    lines.append("- Results sensitive to preprocessing; careless choices can materially change outcomes.\n")

    # Sensitivity summary
    lines.append("## Sensitivity Summary\n")
    lines.append(assumptions.to_markdown(index=False))
    lines.append("\n\n**Interpretation:**\n")
    lines.append("- Median imputation slightly lowers return (0.12 â†’ 0.10).\n")
    lines.append("- Outlier removal improves return (0.12 â†’ 0.14).\n")

    # Decision implications section 
    lines.append("## Decision Implications\n") 
    lines.append("- Use median imputation for reliability.\n")
    lines.append("- Outlier removal helps, but balance risk of discarding stress data.\n")
    lines.append("- Model results shift with data assumptions, so governance on preprocessing is \
    essential before production use.\n") 

    return "\n".join(lines)

summary_md = autodraft_report(assumptions)
with open('../deliverables/final_report.md', 'w', encoding='utf-8') as f:
    f.write(summary_md)
summary_md[:600]
```

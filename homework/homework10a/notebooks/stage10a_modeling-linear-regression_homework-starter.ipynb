## 2) Baseline model fit
Follow the steps; replace `df` with your project DataFrame when ready.

```python
X = df[['vix', 'dgs10', 'fedfunds', 'cpi', 'unrate', 'rv_21d_pct', 'vrp_ratio', 'rate_shock_5d_bps']]
y = df['spx_ret']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)
r2 = r2_score(y_test, y_pred)
# rmse = mean_squared_error(y_test, y_pred, squared=False)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Baseline   R²={r2:.4f}  RMSE={rmse:.6f}')
```

```python
resid = y_test - y_pred
fitted = y_pred
plt.figure(); plt.scatter(fitted, resid); plt.axhline(0, ls='--'); plt.title('Residuals vs Fitted'); plt.show()
plt.figure(); plt.hist(resid, bins=20); plt.title('Residual Histogram'); plt.show()
plt.figure(); st.probplot(resid, dist='norm', plot=plt); plt.title('QQ Plot'); plt.show()
```

## 3) Optional: Add a transformed feature
Add a polynomial or interaction if diagnostics suggest mis-specification.

```python
df['vix_sq'] = df['vix'] ** 2
X2 = df[['vix', 'dgs10', 'fedfunds', 'cpi', 'unrate', 
         'rv_21d_pct', 'vrp_ratio', 'rate_shock_5d_bps', 'vix_sq']]
X2_train, X2_test = X2.iloc[:len(X_train)], X2.iloc[len(X_train):]
lr2 = LinearRegression().fit(X2_train, y_train)
y_pred2 = lr2.predict(X2_test)
mse_2 = mean_squared_error(y_test, y_pred2)
rmse_2 = np.sqrt(mse_2)
r2_2 = r2_score(y_test, y_pred2)
print(f'With vix^2   R²={r2_2:.4f}  RMSE={rmse_2:.6f}')
```

## 4) Interpretation 
- Linearity: The Residuals vs Fitted plot shows that the dots are not randomly scatter around 0, which indicates violation of linearity. The linear model does not fully capture the relationship and potential nonlinear effects remain. 
- Homoscedasticity: The Residuals vs Fitted plot shows a funnel-like pattern which indicates heteroscedasticity. 
- Normality: The histogram shows slight right skew and some large outliers. The points in the Q-Q plot are not following the 45° line and confirms heavy right tail; the deviation indicates non-normality. 
- Independence: Since this is time series financial data, autocorrelation is very likely. 
- Which model do you trust and why?: The baseline model is slightly better since R^2 of baseline is higher than with vix^2 and RMSE of baseline is smaller. Trust the baseline model more, but with caution. It has negative R^2, suggesting the predictors in their current form are not good at explaining short-term S&P returns.

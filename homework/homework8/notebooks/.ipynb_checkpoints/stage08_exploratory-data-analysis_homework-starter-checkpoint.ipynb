```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
sns.set(context='talk', style='whitegrid')
np.random.seed(8)
pd.set_option('display.max_columns', 100)

# === Synthetic data generator (adapt or replace with your own data) ===
n = 160
df = pd.DataFrame({
    'date': pd.date_range('2021-02-01', periods=n, freq='D'),
    'region': np.random.choice(['North','South','East','West'], size=n),
    'age': np.random.normal(40, 8, size=n).clip(22, 70).round(1),
    'income': np.random.lognormal(mean=10.6, sigma=0.3, size=n).round(2),
    'transactions': np.random.poisson(lam=3, size=n),
})
base = df['income'] * 0.0015 + df['transactions']*18 + np.random.normal(0, 40, size=n)
df['spend'] = np.maximum(0, base).round(2)

# inject a bit of missingness and outliers
df.loc[np.random.choice(df.index, 5, replace=False), 'income'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'spend'] = np.nan
df.loc[np.random.choice(df.index, 2, replace=False), 'transactions'] = df['transactions'].max()+12
df.head()
```

## 1) First look

```python
df.info(), df.isna().sum()
```

## 2) Numeric profile

```python
desc = df[['age','income','transactions','spend']].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
desc
```

## 3) Distributions 

```python
fig, axes = plt.subplots(2,2, figsize=(18,10))

# histogram + KDE for a key variable
sns.histplot(df['income'], kde=True, ax=axes[0,0])
axes[0,0].set_title('Income Distribution')

sns.histplot(df['spend'], kde=True, ax=axes[0,1]) 
axes[0,1].set_title('Spend Distribution')

# Tboxplot for outliers
sns.boxplot(x=df['transactions'], ax=axes[1,0])
axes[1,0].set_title('Transactions (Outliers)')

sns.boxplot(x=df['age'], ax=axes[1,1])
axes[1,1].set_title('Age')

plt.tight_layout()
plt.show()
```

## 4) Relationships 

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4))

sns.scatterplot(data=df, x='income', y='spend', hue='region', ax=axes[0])
axes[0].set_title('Income vs Spend')

sns.scatterplot(data=df, x='age', y='spend', ax=axes[1])
axes[1].set_title('Age vs Spend')

# Remove outliers in 'transactions'
df_filtered = df[df['transactions'] <= df['transactions'].quantile(0.99)]

sns.scatterplot(
    data=df_filtered,
    x='transactions', y='spend',
    size='age', sizes=(20, 200),  # scale bubble sizes
    color='steelblue', alpha=0.4, ax=axes[2], legend=False
)
axes[2].set_title('Transactions vs Spend (Size = Age)')

plt.tight_layout()
plt.show()
```

## 5) (Optional) Correlation matrix

```python
corr = df[['age','income','transactions','spend']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
corr
```

## 6) Insights & Assumptions (write your commentary)

### Top 3 insights
- A few customers with very high income and high spend, showing that both income and spend are right-skewed; consider log transforms for linear models. 
- Outliers exist in transactions (>8) and it may distort overall trend. 
- Income and spend are positively correlated; higher income, higher spend. 

### Assumptions & risks 
- Assumes missing values are random (not systematically related). 
- Assumes outliers are genuine customer behavior instead of data entry errors.
- Distributions suggest non-normality and it could affect future modeling if not transformed.

### Next steps before modeling (cleaning & features) 
- Impute missing values (median for income, mean/median for spend).
- Treat outliers by using winsorization or filtering method (especially for transactions).
- Consider log-transforming income and spend to normalize skew.
- Create engineered features like spend per transaction and spend/income ratio.
